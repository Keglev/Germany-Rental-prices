# 4 Cleaning task. Dealing with missing data and outliers. Also drop columns when needed.

```python
# dealing with missing values
def missing_values(df_immo,norows):   # input by the df and the number of rows that you want to show
    total = df_immo.isnull().sum().sort_values(ascending=False)
    percent = ((df_immo.isnull().sum().sort_values(ascending=False)/df_immo.shape[0])*100).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return(missing_data.head(norows))
```

# 5 dealing with incorrect data (NA, 0 values and such)

```python
df_immo['condition'].fillna("Other", inplace=True) # fill the NA by Other
df_immo['condition'].value_counts()
```

```python
# The last 3 is not the good condition for the apartmentt so it will group in 'Other'
#Fill NA of 'yearConstructed' with the mean of each type of condition 'condition' because from my perspective if 
# the apartment is not fully_renovated or refurbished it means that it should have a lot of usage yea
otherscondition = df_immo['condition'].value_counts().tail(3).index

othersregion = list(df_immo['condition'].value_counts().tail(3).index)
def editcondition(df_immolist):
    if df_immolist in otherscondition:
        return 'Other'
    else:
        return df_immolist

df_immo['condition'] =df_immo['condition'].apply(editcondition)
df_immo['condition'].value_counts()
```

```python
#Fill NA of 'yearConstructed' with the mean of each type of condition 'condition' because from my perspective if 
# the apartment is not fully_renovated or refurbished it means that it should have a lot of usage year
df_immo["yearConstructed"] = df_immo['yearConstructed'].fillna(df_immo.groupby('condition')['yearConstructed'].transform('mean')).round(0)
```

```python
#Create the new variables to tell the duration since last renovated or built until today
df_immo['numberOfYear'] = date.today().year - df_immo["yearConstructed"]
```

```python
# fill all the missing values
df_immo = df_immo[(df_immo['floor'] >= -1) & (df_immo['floor'] <= 20)] # Floor should contain only basement - 20th floor
df_immo['heatingType'].fillna(df_immo['heatingType'].mode()[0], inplace=True)
df_immo['typeOfFlat'].fillna(df_immo['typeOfFlat'].mode()[0], inplace=True)
```

```python
heatinglist = list(df_immo['heatingType'].value_counts().head(10).index)
df_immo = df_immo[pd.DataFrame(df_immo['heatingType'].tolist()).isin(heatinglist).any(1).values]
```

```python
df_immo.head()
```

```python
for cols in df_immo.columns:
    if df_immo[cols].dtype == 'int64' or df_immo[cols].dtype == 'float64':
        upper_range = df_immo[cols].mean() + 3 * df_immo[cols].std()
        lower_range = df_immo[cols].mean() - 3 * df_immo[cols].std()
        
        indexs = df_immo[(df_immo[cols] > upper_range) | (df_immo[cols] < lower_range)].index
        df_immo = df_immo.drop(indexs)
```
